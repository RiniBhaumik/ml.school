```python
lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))
tokenizer = RegexpTokenizer(r'\w+')  # Advanced tokenizer that removes punctuation

def advanced_text_processing(text):
    # Convert to lowercase
    text = str(text).lower()

    # Tokenize with advanced tokenizer
    tokens = tokenizer.tokenize(text)

    # Remove stopwords and short tokens
    tokens = [word for word in tokens if word not in stop_words and len(word) > 2]

    # Lemmatization
    tokens = [lemmatizer.lemmatize(word) for word in tokens]

    return ' '.join(tokens)

# Apply text processing
df['proc_text'] = df['text'].apply(advanced_text_processing)
# test_df['proc_text'] = test_df['text'].apply(advanced_text_processing)
```
